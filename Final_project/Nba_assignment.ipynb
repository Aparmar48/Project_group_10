```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
nba_data = pd.read_csv("C:/Users/agp07/Project_group_10/Datasets/all_seasons.csv")
nba_data
```

# Top players in each season:

```python
# Top players in each season and the points they have scored, rebounds and ts_pct (shooting efficiency of player that takes
# 2 and 3 point shots)
top_players = nba_data[["player_name", "pts", "season","reb", "ts_pct", "player_height", "player_weight", "country"]]
players_by_season = top_players.loc[top_players.groupby(top_players["season"])["pts"].idxmax()]
players_by_season
```

```python
# Finding correlation and regression in between Ave_height and Ave_weight
Ave_hgt = players_by_season["player_height"]
Ave_wgt = players_by_season["player_weight"]
correlation = st.pearsonr(Ave_hgt, Ave_wgt)
print(f'The correlation between player height and weight is {round(correlation[0],2)}')
```

```python
from scipy.stats import linregress
```

```python
(slope, intercept, rvalue, pvalue, stderr)= linregress(Ave_hgt, Ave_wgt)
regress_values = Ave_hgt * slope + intercept

plt.scatter(Ave_hgt, Ave_wgt, color="b", alpha=0.5)
plt.plot(Ave_hgt, regress_values, color='red')
plt.xlabel("Ave_height (Cm)")
plt.ylabel("Ave_weight (kg)")
plt.title("Players Height vs. Weight", fontsize=17)
plt.savefig("C:/Users/agp07/OneDrive/Pictures/Saved 1 Pictures")
plt.show()
```

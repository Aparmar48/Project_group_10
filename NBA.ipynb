```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Resources/all_seasons.csv', index_col='Unnamed: 0')
df = df.sort_values(by='season').reset_index(drop=True)
df.head()
```

```python
country_count = df["country"].value_counts()
final_data = country_count
final_data
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
 
if __name__ == '__main__':
    # creating data on which bar chart will be plot
    x = ["USA", "France", "Canada",
         "Brazil", "Australia"]
    y = [9836, 167, 161, 83,
         83]
     
    # making the bar chart on the data
    plt.bar(x, y)
     
    # calling the function to add value labels
    addlabels(x, y)
     
    # giving title to the plot
    plt.title("Top 5 Countries have most player")
     
    # giving X and Y labels
    plt.xlabel("Country")
    plt.ylabel("Number of Players")
     
    # visualizing the plot
    plt.show()
```

```python
grouped_data = df.groupby(["season", "country"])["player_name"].count()
final_data1 = grouped_data.reset_index()
final_data1
```

```python
#Number of players by country season 1996-97
df_1996 = final_data1[ final_data1['season'] == "1996-97"]
newdf1996 = df_1996.sort_values(by="player_name", ascending=False)
newdf1996
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
 
if __name__ == '__main__':
    # creating data on which bar chart will be plot
    x = ["Canada", "Congo", "Croatia",
         "Jamaica", "Lithuania", "Nigeria", "Serbia", "Ukraine"]
    y = [2, 1, 1, 1, 1, 1, 1, 1]
    plt.bar(x, y)
    addlabels(x, y)
    plt.title("Players by Country season 1996-97")
    plt.xlabel("Country")
    plt.ylabel("Number of Players")
    plt.xticks(rotation=45)
    # visualizing the plot
    plt.show()
```
